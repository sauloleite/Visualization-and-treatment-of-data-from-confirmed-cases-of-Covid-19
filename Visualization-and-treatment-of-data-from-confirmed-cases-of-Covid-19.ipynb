# Casos confirmados de COVID-19

```python
import pandas
from urllib.request import urlretrieve
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
urlretrieve(url, 'datasets/goblal_cases_covid19.csv')
```

```python
df_covid = pandas.read_csv('datasets/goblal_cases_covid19.csv')
```

```python
df_covid
```

```python
df_covid.drop(['Lat', 'Long'], axis=1, inplace=True)
```

```python
df_covid
```

```python
df_country = df_covid.groupby('Country/Region').sum()
```

```python
df_country
```

```python
df_country.loc['Brazil']
```

```python
date = df_country.loc['Brazil'].index
cases = df_country.loc['Brazil'].values
```

```python
date
```

```python
cases
```

```python
plt.figure(figsize=(14,8))
plt.bar(date, cases)
```

```python
s_brazil = df_country.loc['Brazil']
```

```python
s_brazil = s_brazil[s_brazil > 0]
```

```python
s_brazil
```

```python
s_brazil.values
```

```python
tam = len(s_brazil)
plt.figure(figsize=(16,8))
plt.xticks(rotation = 60)
plt.bar(s_brazil.index[tam-60:tam], s_brazil.values[tam-60:tam])
plt.title('Total de casos confirmados de COVID-19 no Brasil')
```
